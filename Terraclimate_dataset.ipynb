```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("TerraClimate_AllBands_BD_1958_2024.csv")
df.head()
```

```python
import ast
# Drop 'system:index' column
df = df.drop(columns=['system:index'])

# Extract 'latitude' and 'longitude' from '.geo' JSON
df[['longitude', 'latitude']] = df['.geo'].apply(lambda geo: pd.Series(ast.literal_eval(geo)['coordinates']))

# Drop the original '.geo' column
df = df.drop(columns=['.geo'])
```

```python
df.head()
```

```python
first_cols = ['name', 'longitude', 'latitude', 'month', 'year']
# Then add the rest (excluding duplicates)
other_cols = [col for col in df.columns if col not in first_cols]
df = df[first_cols + other_cols]

df.head()
```

```python
df.to_csv("TerraClimate_Dataset_ 1958_2024_cleaned.csv", index=False)
```
